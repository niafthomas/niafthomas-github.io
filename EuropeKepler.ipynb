Bringing in the social mobility and home ownership data from World Economic Forum and Wikipedia to match it to the country names. (Previously scraped this data)

```python
data=pd.read_csv('https://raw.githubusercontent.com/niafthomas/niafthomas.github.io/main/Project/Data/Europe_HomeO_SM.csv').set_index('Country')
```

```python
data
```

```python
colour_code = [1,2,3,4,5,6,7,8,9]
```

```python
# conditions to assign colour value based on combination of rate of home ownership and social mobility
conditions = [
    (data['Home ownership rate(%)'] < 68) & (data['Index Score'] < 69),
    (data['Home ownership rate(%)'] < 68) & (data['Index Score'] > 69) & (data['Index Score'] < 76),
    (data['Home ownership rate(%)'] < 68) & (data['Index Score'] > 76),
    (data['Home ownership rate(%)'] > 68) & (data['Home ownership rate(%)'] < 80) & (data['Index Score'] < 69),
    (data['Home ownership rate(%)'] > 68) & (data['Home ownership rate(%)'] < 80) & (data['Index Score'] > 69) & (data['Index Score'] < 76),
    (data['Home ownership rate(%)'] > 68) & (data['Home ownership rate(%)'] < 80) & (data['Index Score'] > 76),
    (data['Home ownership rate(%)'] > 68) & (data['Index Score'] < 59),
    (data['Home ownership rate(%)'] > 68) & (data['Index Score'] > 59) & (data['Index Score'] < 76),
    (data['Home ownership rate(%)'] > 68) & (data['Index Score'] > 76)
    ]
```

```python
# create a list of the values we want to assign for each condition
colour_code = [1,2,3,4,5,6,7,8,9]

# create a new column and use np.select to assign values to it using our lists as arguments
data["Colour"] = np.select(conditions, colour_code)
```

```python
data["Colour"] = np.select(conditions, colour_code)
data
```

We extend the topojson properties

```python
for i in topo['objects']['continent_Europe_subunits']['geometries']:
  country=i['properties']['geounit']
  if country in data.index:
    print(country,'found! Extending data..')
    i['properties']['Scoial Mobility Score']=data.loc[country]['Index Score']
    i['properties']['Home Ownership Rate (%)']=data.loc[country]['Home ownership rate(%)']
    i['properties']['Colour']=data.loc[country]['Colour']
```
