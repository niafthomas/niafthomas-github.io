```python
tables=pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_income_equality')
```

```python
tables[0]
```

```python
df2=tables[0]
df2.columns = ["Country", "a", "b", "c", "a", "Gini", "b", "c", "a", "b", "c"]
df2
```

```python
df2 = df2.drop(["a", "b", "c"], axis = 1)
df2
```

```python
gini = df2
```

```python
gini = gini.dropna()
gini
```

```python
gini.to_csv("gini.csv", index = False)
from google.colab import files
files.download("gini.csv")
```

```python
sm_gini = pd.merge(gini, sm, how='outer', on='Country')
```

```python
sm_gini
```

```python
sm_gini = sm_gini.dropna()
```

```python
sm_gini
```

```python
sm_gini.to_csv("sm_gini.csv", index = False)
```
