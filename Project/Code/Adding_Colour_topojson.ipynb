Bringing in the Social Mobility and Gini Coefficient data from World Economic Forum and World Bank to match it to the country names.

```python
import pandas as pd
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/niafthomas/niafthomas.github.io/main/sm_gini.csv').set_index('Country')
```

```python
data
```

```python
colour_code = [1,2,3,4,5,6,7,8,9]
```

```python
data
```

```python

# conditions to assign colour value based on combination of gdp and happiness
conditions = [
    (data['Gini'] < 33) & (data['Index Score'] < 54),
    (data['Gini'] < 33) & (data['Index Score'] > 54) & (data['Index Score'] < 69),
    (data['Gini'] < 33) & (data['Index Score'] > 69),
    (data['Gini'] > 33) & (data['Gini'] < 39) & (data['Index Score'] < 54),
    (data['Gini'] > 33) & (data['Gini'] < 39) & (data['Index Score'] > 54) & (data['Index Score'] < 69),
    (data['Gini'] > 33) & (data['Gini'] < 39) & (data['Index Score'] > 69),
    (data['Gini'] > 39) & (data['Index Score'] < 54),
    (data['Gini'] > 39) & (data['Index Score'] > 54) & (data['Index Score'] < 69),
    (data['Gini'] > 39) & (data['Index Score'] > 69)
    ]
```

```python
# create a list of the values we want to assign for each condition
colour_code = [1,2,3,4,5,6,7,8,9]

# create a new column and use np.select to assign values to it using our lists as arguments
data["Colour"] = np.select(conditions, colour_code)
```

```python
data["Colour"] = np.select(conditions, colour_code)
```

```python
data
```

We extend the topojson properties - in your final product, you need to rename countries until you have a perfect match.

```python
for i in topo['objects']['countries1']['geometries']:
  country=i['properties']['name']
  if country in data.index:
    print(country,'found! Extending data..')
    i['properties']['Scoial Mobility Score']=data.loc[country]['Index Score']
    i['properties']['Gini Coefficient']=data.loc[country]['Gini']
    i['properties']['Colour']=data.loc[country]['Colour']
```
