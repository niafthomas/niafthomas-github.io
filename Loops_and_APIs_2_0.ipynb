```python
import requests
import os
import json
from google.colab import files
```

```python
# Data to download from Federal Reserve Economic Data (FRED):
# 1) Market Yield on U.S. Treasury Securities at 10-Year Constant Maturity (DGS10)
# 2) Japanese Yen to U.S. Dollar Spot Exchange Rate (DEXJPUS)
# 3) S&P 500 (SP500)	
# 4) Consumer Price Index: Total All Items for the United States (CPALTT01USM659N)

# unique API key: b2abc44b6fb81c0623d89b7bba0ae907
url_base = "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=b2abc44b6fb81c0623d89b7bba0ae907&file_type=json"

# File base name:
file_base = "data-Fred-{}.json"

## Series:
fredSeries = ['DGS10', 'DEXJPUS', 'SP500', 'CPALTT01USM659N']

## Print URLs in a loop
for i in fredSeries:
  print(i)

  # Build URL for iteration of the loop
  URL = url_base.format(i)
  print(URL)

  #Request the html from the URL
  data = requests.get(URL).json()
  print(data)

  #Set the filename
  fileName = file_base.format(i)
  print(fileName)

  #Save the file
  with open(fileName, 'w', encoding='utf-8') as f:
     json.dump(data, f, ensure_ascii=False, indent=4)
  
  #Download the file to local machine
  files.download(fileName)
```
